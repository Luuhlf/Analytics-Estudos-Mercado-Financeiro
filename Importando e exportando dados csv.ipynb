```python
import yfinance as yf
```

```python
msft = yf.download('MSFT', start = '2021-01-01')
```

```python
msft.head()
```

```python
msft_ohlc = msft[['Open', 'High', 'Low', 'Close']]
```

```python
msft_ohlc.head()
```

```python
msft_ohlc.plot()
```

```python
msft_ohlc['Close'].plot();
```

```python
abev = yf.download('ABEV3.SA', start = '2021-01-01')
```

```python
abev.head()
```

```python
abev_ohlc = abev[['Open', 'High', 'Low', 'Close']]
```

```python
abev_ohlc
```

```python
abev_ohlc.to_csv('ABEV3_OHLC.csv', sep = ';', decimal = ',')
```
